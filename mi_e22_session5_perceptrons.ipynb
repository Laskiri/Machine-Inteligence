```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets

X, y = datasets.make_blobs(n_samples=150,n_features=2,
                           centers=2,cluster_std=1.05,
                           random_state=2)



#Plotting
fig = plt.figure(figsize=(10,8))
plt.plot(X[:, 0][y == 0], X[:, 1][y == 0], 'r^')
plt.plot(X[:, 0][y == 1], X[:, 1][y == 1], 'bs')
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.title('Random Classification Data with 2 classes')
```

```python
def decision(x, w, b):
    return np.dot(x,w) + b > 0
```

```python
def perceptron(X, y, lr, epochs):
    m, n = X.shape

    w = np.zeros(n)
    b = 0

    # Training.
    for epoch in range(epochs):

        # looping for every example.
        for idx, x_i in enumerate(X):
            d = decision(x_i, w, b)

            if d == y[idx]:
              continue

            # IMPLEMENT UPDATES HERE
            # Both the bias (b) and the weight (w) should be updated according to the pseudocode



    return w, b
```

```python
def plot_decision_boundary(X, theta, b):
    """Plotting magic :)"""
    x1 = [min(X[:,0]), max(X[:,0])]
    m = -theta[0] / theta[1]
    c = -b / theta[1]
    x2 = x1 * np.asarray([m], dtype=np.float64) + c

    # Plotting
    fig = plt.figure(figsize=(10,8))
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "r^")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs")
    plt.xlabel("feature 1")
    plt.ylabel("feature 2")
    plt.title('Perceptron Algorithm')
    plt.plot(x1, x2, 'y-')
```

```python
w, b = perceptron(X, y, 0.5, 100)
plot_decision_boundary(X, w, b)
```
